```python
import google.colab
import pathlib

mount_path = "/content/drive/"
data_root = "Colab Notebooks/BMI706/Project/NHANE_1314"
data_root = pathlib.Path(data_root)
data_root = pathlib.Path(mount_path) / 'MyDrive' / data_root
google.colab.drive.mount(mount_path)
```

```python
questionnaire_path = pathlib.Path(data_root) / 'questionnaire.csv'
examination_path = pathlib.Path(data_root) / 'examination.csv'
medications_path = pathlib.Path(data_root) / 'medications.csv'
labs_path = pathlib.Path(data_root) / 'labs.csv'
diet_path = pathlib.Path(data_root) / 'diet.csv'
demographic_path = pathlib.Path(data_root) / 'demographic.csv'
```

```python
import pandas as pd

questionnaire_df = pd.read_csv(questionnaire_path)
examination_df = pd.read_csv(examination_path)
#medications_df = pd.read_csv(medications_path, encoding_errors='ignore')
labs_df = pd.read_csv(labs_path)
diet_df = pd.read_csv(diet_path)
demographic_df = pd.read_csv(demographic_path)

df_list = [questionnaire_df, examination_df, labs_df, diet_df, demographic_df]
merged_df = df_list[0]
for df in df_list[1:]:
    merged_df = merged_df.merge(df, on='SEQN', how='inner')
```

```python
var_list = ['BMXBMI', 'RIAGENDR', 'RIDRETH3', 'RIDAGEYR', 'INDFMPIR', 'LBDLDL', 'BPXSY1', 'DIQ010']
merged_clean_df = merged_df[var_list].dropna()
merged_clean_df = merged_clean_df.rename(columns={'BMXBMI': 'BMI', 'RIAGENDR': 'Gender', 'RIDRETH3': 'Race', 'RIDAGEYR': 'Age', 'INDFMPIR': 'Income Ratio', 'LBDLDL': 'LDL', 'BPXSY1': 'Blood Pressure', 'DIQ010': 'Diabetes'})
```

```python
save_path = pathlib.Path(data_root) / 'merged_clean_data.csv'
merged_clean_df.to_csv(save_path, index=False)
```
